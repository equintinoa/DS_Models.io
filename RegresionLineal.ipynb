## Caso Bidimensional del Modelo

Con el objetivo de entender mejor las justificaciones del modelo, comenzaremos por abordar un caso sencillo de bajas dimensiones y posteriormente generalizaremos las ideas. Supongamos que, dadas dos observaciones de una variable de entrada, queremos predecir los valores de una variable objetivo. Usando el modelo de regresión lineal, esto se representa como:

$$
\begin{cases}
\hat{y}_1=\hat{\beta}_1 x_1+\hat{\beta}_2 \\
\hat{y}_2=\hat{\beta}_1 x_2+\hat{\beta}_2
\end{cases}
$$

Para futuros cálculos, es conveniente representar al modelo en forma matricial; por lo tanto definimos las siguientes matrices

$$
\boldsymbol{X}=\begin{bmatrix}
x_{11} & x_{12} \\
x_{21} & x_{22}
\end{bmatrix}=\begin{bmatrix}
x_1 & 1 \\
x_2 & 1
\end{bmatrix} ; \quad \boldsymbol{\hat{Y}}=\begin{bmatrix}
\hat{y}_1 \\
\hat{y}_2
\end{bmatrix} ; \quad \boldsymbol{\hat{\beta}}=\begin{bmatrix}
\hat{\beta}_1 \\
\hat{\beta}_2
\end{bmatrix}
$$

De esta manera, el modelo se representa como:

$$
\boldsymbol{\hat{Y}} = \boldsymbol{X} \boldsymbol{\hat{\beta}}
$$


<details>
<summary>Detalle</summary>

$$
\begin{aligned}
\left[\begin{array}{l}
\hat{y}_1 \\
\hat{y}_2
\end{array}\right] &= \left[\begin{array}{ll}
1 & x_1 \\
1 & x_2
\end{array}\right]\left[\begin{array}{l}
\hat{\beta}_1 \\
\hat{\beta}_2
\end{array}\right] \\
&= \left[\begin{array}{l}
\hat{\beta}_1 x_1 +\hat{\beta}_2 \\
\hat{\beta}_1 x_2  + \hat{\beta}_2
\end{array}\right]
\end{aligned}
$$

</details>


### Optimización del Modelo

La pregunta que surge naturalmente es: ¿Cómo podemos escoger los coeficientes adecuados  $\beta _1$ y $\beta _2$? Para esto definimos la función “Suma de los Cuadrados Residuales” como
$$
\begin{aligned}
\operatorname{RSS}(\boldsymbol{\hat{\beta}}) & =\left(y_{1}-\hat{y}_{1}\right)^{2}+\left(y_{2}-\hat{y}_{2}\right)^{2}=\sum_{i=1}^{2}\left(y_{i}-\hat{y}_{i}\right)^{2} \\
\end{aligned}
$$

Esta medida de error puede representarse de forma matricial como

$$
\operatorname{RSS}(\boldsymbol{\hat{\beta}})= \left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}}\right)^{T} \left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}} \right)
$$

<details>
<summary>Detalle</summary>

$\begin{aligned}\left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}}\right)^{T} \left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}} \right) & =\left[\begin{array}{c}y_1-\left(\hat{\beta}_1 x_n+\hat{\beta}_2 x_{12}\right) \\ y_2-\left(\hat{\beta}_1 x_{21}+\hat{\beta}_2 x_{22}\right)\end{array}\right]^{\top}\left[\begin{array}{c}y_1-\left(\hat{\beta}_1 x_n+\hat{\beta}_2 x_{12}\right) \\ y_2-\left(\hat{\beta}_1 x_{21}+\hat{\beta}_2 x_{22}\right)\end{array}\right] \\
 & =\left(y_1-\left(\hat{\beta}_1 x_{11}+\hat{\beta}_2 x_{12}\right)\right)^2+\left(y_2-\left(\hat{\beta}_1 x_{21}+\beta_2 x_{22}\right)^2\right) \\ & =\left(y_1-\hat{y}_1\right)^2+\left(y_2-\hat{y}_2\right)^2\end{aligned}$

</details>

Para encontrar los mejores valores en los parámetros, diferenciamos lo anterior

$$
\begin{aligned}
\frac{\partial}{\partial \hat{\beta}_1} \operatorname{RSS}(\boldsymbol{\hat{\beta}}) & =-2\left(y_1-\left(\hat{\beta}_1 x_{11}+\hat{\beta}_2 x_{12}\right)\right) x_{11}-2\left(y_2-\left(\hat{\beta}_1 x_{21}+\hat{\beta}_2 x_{22}\right)\right) x_{21} \\
& =-2 \sum_{i=1}^2\left(y_i-\sum_{j=1}^2 \hat{\beta}_j x_{i j}\right) x_{i 1} \\
\frac{\partial}{\partial \hat{\beta}_2} \operatorname{RSS}(\boldsymbol{\hat{\beta}}) & =-2\left(y_1-\left(\hat{\beta}_1 x_{11}+\hat{\beta}_2 x_{12}\right)\right) x_{12}-2\left(y_2-\left(\hat{\beta}_1 x_{21}+\hat{\beta}_2 x_{22}\right)\right) x_{22} \\
& =-2 \sum_{i=1}^2\left(y_i-\sum_{j=1}^2 x_{i j} \hat{\beta}_j\right) x_{i 2}
\end{aligned}
$$

De esta forma se tiene

$$
    \frac{\partial}{\partial \boldsymbol{\hat{\beta}}} \operatorname{RSS}(\boldsymbol{\hat{\beta}})=\left[\begin{array}{l}
    -2 \sum_{i=1}^2\left(y_i-\sum_{j=1}^2 x_{i j} \hat{\beta}_j\right) x_{i 1} \\
    -2 \sum_{i=1}^2\left(y_i-\sum_{j=1}^2 x_{i j} \hat{\beta}_j\right) x_{i 2}
    \end{array}\right]
$$

Lo cual se puede representar como

$$
    \frac{\partial}{\partial \boldsymbol{\hat{\beta}}} \operatorname{RSS}(\boldsymbol{\hat{\beta}}) = -2 \boldsymbol{X}^{T} \left( \boldsymbol{Y} - \boldsymbol{X \hat{\beta}} \right)
$$

<details>
<summary>Detalle</summary>

$$
\begin{aligned}
-2 \boldsymbol{X}^{T} \left( \boldsymbol{Y} - \boldsymbol{X \hat{\beta}} \right) & =-2\left[\begin{array}{ll}
x_{11} & x_{21} \\
x_{12} & x_{22}
\end{array}\right]\left[\begin{array}{l}
y_1-\left(\hat{\beta}_1 x_{11}+\hat{\beta}_2 x_{12}\right) \\
y_2-\left(\hat{\beta}_2 x_{21}+\hat{\beta}_2 x_{22}\right)
\end{array}\right] \\
& =-2\left[\begin{array}{l}
\left(y_1-\left(\hat{\beta}_1 x_n+\hat{\beta}_2 x_{12}\right)\right) x_{11}+\left(y_2-\left(\hat{\beta}_1 x_{21}+\hat{\beta}_2 x_{22}\right)\right) x_{21} \\
\left(y_2-\left(\hat{\beta}_1 x_{11}+\hat{\beta}_2 x_{12}\right)\right) x_{12}+\left(y_2-\left(\hat{\beta}_1 x_{21}+\beta_2 x_{22}\right)\right) x_{22}
\end{array}\right]
\end{aligned}
$$
</details>

Tenemos la siguiente ecuación normal, la misma tiene solución si $\boldsymbol{X^{T} X}$ es no singular.
$$
2 \boldsymbol{X}^{T} \left( \boldsymbol{Y} - \boldsymbol{X \hat{\beta}} \right)= \boldsymbol{0} \Rightarrow  \boldsymbol{ \hat{\beta}} = \left( \boldsymbol{X}^{T} \boldsymbol{X} \right)^{-1} \boldsymbol{X}^{T} \boldsymbol{Y}
$$

## Generalización del modelo

Para obtener una generalización del modelo, supongamos que tenemos $n$ observaciones de $m$ variables. Así tenemos las siguientes matrices.

$$
\boldsymbol{Y} =\left[\begin{array}{c}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{array}\right] ; \quad \boldsymbol{X}=\left[\begin{array}{cccc}
x_{11} & x_{12} & \cdots & x_{1 m} \\
x_{21} & x_{22} & \cdots & x_{2 m} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n 1} & x_{m 2} & \cdots & x_{n m}
\end{array}\right] ; \quad \boldsymbol{{\beta}}=\left[\begin{array}{c}
{\beta}_1 \\
{\beta}_2 \\
\vdots \\
{\beta}_m
\end{array}\right]
$$

Sin pérdida de generalidad, supongamos que  $x_{1m}, x_{2m}, \dots, x_{nm}$ son iguales a $1$. De esta forma, el modelo se representa como 

$$
    \boldsymbol{Y} = \boldsymbol{X \beta}
$$

### Suma de los cuadrados residuales $\operatorname{RSS}$

Para generalizar la funcion $\operatorname{RSS}(\hat{\boldsymbol{\beta}})$ se tiene

$$
    \begin{aligned}
    \operatorname{RSS}(\boldsymbol{\hat{\beta}}) & =\left(y_1-\hat{y}_1\right)^2+\left(y_2-\hat{y}_2\right)^2+\cdots+\left(y_n-\hat{y}_n\right)^2 \\
    & =\sum_{k=1}^n\left(y_k-\hat{y}_k\right)^2 \\
    & =\sum_{k=1}^n\left(y_k-\sum_{i=1}^m \hat{\beta}_i x_{k i}\right)^2 \\
    & =\left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}}\right)^{T} \left( \boldsymbol{Y} - \boldsymbol{X} \boldsymbol{\hat{\beta}} \right)
    \end{aligned}
$$

Al derivar cada término de la suma anterior obtenemos

$$
-2\left[y_k-\left(\hat{\beta}_1 x_{k n}+\hat{\beta}_2 x_{k 2}+\cdots+\hat{\beta}_j x_{k j}+\cdots+\beta_m x_{k m}\right)\right] x_{k j}=-2\left[y_k-\sum_{i=1} ^{m} \beta_i x_{k i}\right] x_{k j}
$$

Por lo tanto, la derivada de la función $\operatorname{RSS}(\boldsymbol{\hat{\beta}})$ es

$$
\begin{aligned}
& \quad \frac{\partial}{\partial \hat{\beta}_j} \operatorname{RSS}(\boldsymbol{\hat{\beta}})=-2 \sum_{k=1}^n\left(y_k-\sum_{i=1}^m \hat{\beta}_i x_{k i}\right) x_{k j} \\
& \Rightarrow \frac{\partial}{\partial \boldsymbol{\hat{\beta}}} \operatorname{RSS}(\hat{\beta})=-2\left[\begin{array}{c}
\sum_{k=1}^n\left(y_k-\sum_{i=1}^m \hat{\beta}_i x_{k i}\right) x_{k 1} \\
\sum_{k=1}^n\left(y_k-\sum_{i=1}^m \hat{\beta}_i x_{k i}\right) x_{k 2} \\
\vdots \\
\sum_{k=1}^n\left(y_k-\sum \hat{\beta}_i x_{k i}\right) x_{k m}
\end{array}\right] \\
& \therefore \frac{\partial}{\partial \boldsymbol{\hat{\beta}}}  \operatorname{RSS}(\hat{\beta}) =-2 \boldsymbol{X}^{T} \left( \boldsymbol{Y} - \boldsymbol{X \hat{\beta}} \right)
\end{aligned}
$$

Si $\boldsymbol{X^{T} X}$ es no singular, la siguiente ecuación tiene solución.
$$
2 \boldsymbol{X}^{T} \left( \boldsymbol{Y} - \boldsymbol{X \hat{\beta}} \right)= \boldsymbol{0} \Rightarrow  \boldsymbol{ \hat{\beta}} = \left( \boldsymbol{X}^{T} \boldsymbol{X} \right)^{-1} \boldsymbol{X}^{T} \boldsymbol{Y}
$$


## Implementación del modelo en Python.

Para esta sección, utilizaremos las deducciones hechas en el apartado anterior sobre cuál es el parámetro óptimo $\hat{\boldsymbol{\beta}}$ en el modelo de regresión lineal.

```python
#Dependencias
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

from sklearn.model_selection import train_test_split
```

Para mayor claridad, crearemos una función para obtener el parámetro óptimo `coef_LinnearModel()` y otra función para aplicar el modelo a los datos `linear_regression()`. 


```python
def coef_LinnearModel(X, Y):

    X = np.c_[X, np.ones(X.shape[0])]
    X_t = np.transpose(X)
    beta = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_t, X)), X_t), Y)
    
    return beta

def linear_regression(X, beta):
    X = np.c_[X, np.ones(X.shape[0])]
    Y_pred = np.matmul(X, beta)
    return Y_pred
```

### Aplicación del Modelo a un Data Frame
Con el objetivo de probar el funcionamiento del modelo y el código desarrollado, usaremos los datos guardados en el archivo `precios_casas.csv`. Este conjunto de datos contiene 1845 registros sobre el área de construcción y el precio en el mercado de casas.  Como se puede observar en la gráfica, estas dos variables parecen guardar una fuerte correlación, lo cuál las hace ideales para la implementación de nuestro modelo.


```python
casas_df = pd.read_csv('./precios_casas.csv')
casas_df.head(10)
```

```python
plt.figure(figsize=(10, 5), facecolor='#282a36')  # Cambia estos valores según tus necesidades

sns.scatterplot(data=casas_df, x='Metros_Cuadrados', y='Precio', color = '#bd93f9')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.gca().set_facecolor('#282a36')

ax = plt.gca() 
for spine in ax.spines.values():
    spine.set_color('#f8f8f2') 

plt.tick_params(axis='x', colors='#f8f8f2')
plt.tick_params(axis='y', colors='#f8f8f2')

plt.title('Precio de la Vivienda vs Área de Construcción', color = '#f8f8f2')
plt.xlabel('Metros Cuadrados', color = '#f8f8f2')
plt.ylabel('Precio', color = '#f8f8f2')
plt.show()

```

Para realizar predicciones de manera adecuada, dividiremos al dataframe anterior en conjuntos de entrenamiento y de prueba. Así, entrenaremos el modelo con el conjunto de prueba y realizaremos las predicciones bajo el conjunto de pruebas.

```python
X = np.transpose(casas_df.iloc[:, 0].values).reshape(-1, 1) 
y=np.transpose(casas_df.iloc[:, 1].values).reshape(-1, 1) 

#Conjuntos de entrenamiento y testeo 
X_train, X_test , y_train, y_test = train_test_split(
    X, y, test_size= 0.8, random_state=345 
)

beta = coef_LinnearModel(X_train,y_train)
y_pred = linear_regression(X_test, beta)
```

El arreglo `y_pred` contiene nuestras predicciones para el conjunto de prueba. Al evaluar estos dos conjuntos `y_test` y `y_pred` vemos que nuestro modelo se ajusta como una linea recta a nuestros datos, tal como lo esperábamos. 

```python
plt.figure(figsize=(10, 5), facecolor='#282a36')

plt.scatter(X_test, y_test, color = '#bd93f9')

plt.scatter(X_test, y_pred, color = '#6272a4')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.gca().set_facecolor('#282a36')

ax = plt.gca() 
for spine in ax.spines.values():
    spine.set_color('#f8f8f2') 

plt.tick_params(axis='x', colors='#f8f8f2')
plt.tick_params(axis='y', colors='#f8f8f2')

plt.title('Precio de la Vivienda vs Área de Construcción', color = '#f8f8f2')
plt.xlabel('Metros Cuadrados', color = '#f8f8f2')
plt.ylabel('Precio', color = '#f8f8f2')
plt.show()
```
